```python
# Importar los m√≥dulos que sean necesarios
import random  # Para selecciones aleatorias
import time    # Para pausas en el juego de memoria

r = "\033[91m"
g = "\033[92m"
y = "\033[93m"
rst = "\033[0m"

# Juego de triqui
def inicializar_tablero_triqui():
    # Inicializa un tablero 3x3 con celdas vac√≠as (" ") para el juego de triqui
    return [[' ' for _ in range(3)] for _ in range(3)]

def imprimir_tablero_triqui(tablero):
    # Muestra el tablero de triqui con una gu√≠a num√©rica (1-9) y el estado actual del juego
    print("\nSelecciona una celda del 1 al 9 seg√∫n esta gu√≠a, o 'R' para rendirte:")
    print("1 | 2 | 3")
    print("---------")
    print("4 | 5 | 6")
    print("---------")
    print("7 | 8 | 9")
    print("")
    print("Estado actual del juego: ")
    for i, fila in enumerate(tablero):
        print(' | '.join(fila))
        if i < 2:
            print('-' * 9)

def celda_a_coordenadas_triqui(celda):
    # Convierte un n√∫mero de celda (1-9) a coordenadas (fila, columna) para el tablero
    celda = celda - 1
    fila = celda // 3
    columna = celda % 3
    return fila, columna

def coordenadas_a_celda_triqui(fila, columna):
    # Convierte coordenadas (fila, columna) a un n√∫mero de celda (1-9)
    return fila * 3 + columna + 1

def hay_ganador_triqui(tablero, jugador):
    # Verifica si hay un ganador revisando filas, columnas y diagonales
    for i in range(3):
        if all([tablero[i][j] == jugador for j in range(3)]):
            return True
        if all([tablero[j][i] == jugador for j in range(3)]):
            return True
    if all([tablero[i][i] == jugador for i in range(3)]):
        return True
    if all([tablero[i][2 - i] == jugador for i in range(3)]):
        return True
    return False

def tablero_lleno_triqui(tablero):
    # Comprueba si el tablero est√° lleno (sin celdas vac√≠as)
    return all([celda != ' ' for fila in tablero for celda in fila])

def movimientos_disponibles_triqui(tablero):
    # Devuelve una lista de coordenadas de celdas vac√≠as disponibles
    return [(i, j) for i in range(3) for j in range(3) if tablero[i][j] == ' ']

def minimax_triqui(tablero, profundidad, es_maximizador):
    # Implementa el algoritmo Minimax para evaluar el mejor movimiento de la IA
    if hay_ganador_triqui(tablero, 'O'):
        return 10 - profundidad
    if hay_ganador_triqui(tablero, 'X'):
        return profundidad - 10
    if tablero_lleno_triqui(tablero):
        return 0
    if es_maximizador:
        max_eval = -float('inf')
        for mov in movimientos_disponibles_triqui(tablero):
            tablero[mov[0]][mov[1]] = 'O'
            eval = minimax_triqui(tablero, profundidad + 1, False)
            tablero[mov[0]][mov[1]] = ' '
            max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = float('inf')
        for mov in movimientos_disponibles_triqui(tablero):
            tablero[mov[0]][mov[1]] = 'X'
            eval = minimax_triqui(tablero, profundidad + 1, True)
            tablero[mov[0]][mov[1]] = ' '
            min_eval = min(min_eval, eval)
        return min_eval

def movimiento_ia_triqui(tablero):
    # Determina el mejor movimiento para la IA usando Minimax, con 20% de chance de error aleatorio
    mejor_puntaje = -float('inf')
    mejor_mov = None
    for mov in movimientos_disponibles_triqui(tablero):
        tablero[mov[0]][mov[1]] = 'O'
        puntaje = minimax_triqui(tablero, 0, False)
        tablero[mov[0]][mov[1]] = ' '
        if puntaje > mejor_puntaje:
            mejor_puntaje = puntaje
            mejor_mov = mov
    if random.randint(0, 100) < 20:  # 20% de probabilidad de error
        lista_movimientos = movimientos_disponibles_triqui(tablero)
        return random.choice(lista_movimientos) if lista_movimientos else mejor_mov
    return mejor_mov

def jugar_triqui():
    # Ejecuta el juego de triqui, alternando turnos entre usuario (X) y IA (O)
    tablero = inicializar_tablero_triqui()
    usuario = 'X'
    ia = 'O'
    turno_usuario = random.choice([True, False])
    print(g + "\nJUEGO DE TRIQUI" + rst)
    print("Eres 'X'. Selecciona una celda del 1 al 9 seg√∫n la gu√≠a, o 'R' para rendirte.")
    imprimir_tablero_triqui(tablero)
    while True:
        if turno_usuario:
            while True:
                entrada = input("Tu jugada. Ingresa un n√∫mero del 1 al 9 o 'R' para rendirte: ").strip().upper()
                if entrada == 'R':
                    print(r + "Te rendiste. Vuelves al men√∫ principal." + rst)
                    return
                try:
                    celda = int(entrada)
                    if 1 <= celda <= 9:
                        fila, col = celda_a_coordenadas_triqui(celda)
                        if 0 <= fila < 3 and 0 <= col < 3 and tablero[fila][col] == ' ':
                            tablero[fila][col] = usuario
                            break
                        else:
                            print(r + "Celda inv√°lida, ya est√° ocupada o fuera de rango. Intenta de nuevo." + rst)
                    else:
                        print(r + "N√∫mero inv√°lido. Usa un n√∫mero del 1 al 9 o 'R' para rendirte." + rst)
                except ValueError:
                    print(r + "Entrada inv√°lida. Ingresa un n√∫mero, ej: 5, o 'R' para rendirte." + rst)
        else:
            print("Turno de la m√°quina...")
            mov = movimiento_ia_triqui(tablero)
            if mov:
                tablero[mov[0]][mov[1]] = ia
                print(f"La m√°quina juega en la celda: {coordenadas_a_celda_triqui(mov[0], mov[1])}")
        imprimir_tablero_triqui(tablero)
        if hay_ganador_triqui(tablero, usuario):
            print(g + "¬°Ganaste!"+ rst)
            break
        elif hay_ganador_triqui(tablero, ia):
            print(g + "¬°La m√°quina gan√≥!" + rst)
            break
        elif tablero_lleno_triqui(tablero):
            print(y + "¬°Empate!" + rst)
            break
        turno_usuario = not turno_usuario

# Juego de Memoria
def inicializar_tablero_memoria():
    # Inicializa un tablero 6x5 con 15 pares de frutas aleatorias y un tablero visible con "*"
    simbolos = ['üçé', 'üçå', 'üçí', 'üçá', 'üçâ', 'üçì', 'üçë', 'üçç', 'ü•ù', 'üçã', 'ü••', 'üçÖ', 'ü•ë', 'üçÜ', 'ü•ï']
    pares = simbolos * 2
    random.shuffle(pares)
    tablero = [pares[i*5:(i+1)*5] for i in range(6)]
    visible = [['*' for _ in range(5)] for _ in range(6)]
    return tablero, visible

def imprimir_tablero_memoria(visible):
    # Muestra el tablero de Memoria con n√∫meros (1-30) para celdas cubiertas y frutas para las destapadas
    print("\nTablero de memoria. Selecciona dos n√∫meros del 1 al 30 que representen celdas cubiertas, o 'R' para rendirte:")
    for i, fila in enumerate(visible):
        print(f"Fila {i}: {' '.join(str(coordenadas_a_celda_memoria(i, j)) if celda == '*' else celda for j, celda in enumerate(fila))}")

def celda_a_coordenadas_memoria(celda):
    # Convierte un n√∫mero de celda (1-30) a coordenadas (fila, columna) para el tablero de Memoria
    celda = celda - 1
    fila = celda // 5
    columna = celda % 5
    return fila, columna

def coordenadas_a_celda_memoria(fila, columna):
    # Convierte coordenadas (fila, columna) a un n√∫mero de celda (1-30)
    return fila * 5 + columna + 1

def es_par_memoria(tablero, pos1, pos2):
    # Verifica si dos posiciones contienen la misma fruta, excluyendo si son la misma celda
    f1, c1 = pos1
    f2, c2 = pos2
    return tablero[f1][c1] == tablero[f2][c2] and pos1 != pos2

def destapar_y_verificar_memoria(tablero, visible, pos1, pos2):
    # Destapa dos celdas, las muestra por 2 segundos y las tapa si no son par
    f1, c1 = pos1
    f2, c2 = pos2
    visible[f1][c1] = tablero[f1][c1]
    visible[f2][c2] = tablero[f2][c2]
    imprimir_tablero_memoria(visible)
    time.sleep(2)
    if not es_par_memoria(tablero, pos1, pos2):
        visible[f1][c1] = '*'
        visible[f2][c2] = '*'
        return False
    return True

def movimiento_ia_memoria(visible):
    # Elige aleatoriamente dos celdas cubiertas para la IA en el juego de Memoria
    disponibles = [(i, j) for i in range(6) for j in range(5) if visible[i][j] == '*']
    if len(disponibles) < 2:
        return None, None
    pos1 = random.choice(disponibles)
    disponibles.remove(pos1)
    pos2 = random.choice(disponibles)
    return pos1, pos2

def input_usuario_memoria(visible):
    # Recibe y valida la entrada del usuario para seleccionar dos celdas en Memoria
    while True:
        entrada = input("Tu jugada. Ingresa dos n√∫meros del 1 al 30, ej: 1 2, o 'R' para rendirte: ").strip().upper()
        print("\n" * 18)
        if entrada == 'R':
            print(r + "Te rendiste. Vuelves al men√∫ principal." + rst)
            return None, None
        try:
            nums = entrada.split()
            if len(nums) != 2:
                raise ValueError
            celda1, celda2 = map(int, nums)
            if 1 <= celda1 <= 30 and 1 <= celda2 <= 30 and celda1 != celda2:
                pos1 = celda_a_coordenadas_memoria(celda1)
                pos2 = celda_a_coordenadas_memoria(celda2)
                if visible[pos1[0]][pos1[1]] == '*' and visible[pos2[0]][pos2[1]] == '*':
                    return pos1, pos2
                else:
                    print("Celdas inv√°lidas, ya est√°n destapadas. Intenta de nuevo.")
            else:
                print("N√∫meros inv√°lidos. Usa dos n√∫meros del 1 al 30, distintos, o 'R' para rendirte.")
        except ValueError:
            print("Entrada inv√°lida. Ingresa dos n√∫meros, ej: 1 2, o 'R' para rendirte.")

def memoria_terminada(visible):
    # Comprueba si todas las celdas est√°n destapadas (sin "*")
    return all([celda != '*' for fila in visible for celda in fila])

def jugar_memoria():
    # Ejecuta el juego de Memoria, alternando turnos entre usuario y IA
    tablero, visible = inicializar_tablero_memoria()
    puntos_usuario = 0
    puntos_ia = 0
    turno_usuario = random.choice([True, False])
    print(g + "\nJUEGO DE MEMORIA" + rst)
    imprimir_tablero_memoria(visible)
    while not memoria_terminada(visible):
        if turno_usuario:
            pos1, pos2 = input_usuario_memoria(visible)
            if pos1 is None and pos2 is None:
                return
            if destapar_y_verificar_memoria(tablero, visible, pos1, pos2):
                puntos_usuario += 1
                print(g + "¬°Encontraste un par! Sumas 1 punto." + rst)
                print("\n")
            else:
                print(r + "No es un par. Se tapan de nuevo." + rst)
                print("\n")
                imprimir_tablero_memoria(visible)
        else:
            print("\n")
            print(y + "\nTurno de la m√°quina..." + rst)
            pos1, pos2 = movimiento_ia_memoria(visible)
            if pos1 and pos2:
                print(f"La m√°quina elige las celdas {coordenadas_a_celda_memoria(*pos1)} y {coordenadas_a_celda_memoria(*pos2)}")
                if destapar_y_verificar_memoria(tablero, visible, pos1, pos2):
                    puntos_ia += 1
                    print("¬°La m√°quina encontr√≥ un par! Suma 1 punto.")
                else:
                    print(r + "No es un par. Se tapan de nuevo." + rst)
                    print("\n")
                    imprimir_tablero_memoria(visible)
        turno_usuario = not turno_usuario
    print(g + "\n¬°Juego terminado!" + rst)
    print("")
    print(y + f"Tus puntos: {puntos_usuario}" + rst)
    print(r + f"Puntos de la m√°quina: {puntos_ia}" + rst)
    if puntos_usuario > puntos_ia:
        print("")
        print("¬°Ganaste!")
        print("")
    elif puntos_ia > puntos_usuario:
        print("")
        print("¬°La m√°quina gan√≥!")
        print("")
    else:
        print("")
        print("¬°Empate!")
        print("")

def menu_principal():
    # Muestra el men√∫ principal y dirige al juego seleccionado o sale
    while True:
        print(g + "\n¬°Bienvenido a la aplicaci√≥n de juegos contra la m√°quina!" + rst)
        print("")
        print("1. Juego de triqui")
        print("2. Juego de memoria")
        print("3. Salir")
        print("")
        opcion = input("Selecciona un juego: 1, 2, o 3: ").strip()
        print("")
        print("")
        if opcion == '1':
            jugar_triqui()
        elif opcion == '2':
            jugar_memoria()
        elif opcion == '3':
            print("")
            print(g + "¬°Gracias por jugar! Nos vemos." + rst)
            break
        else:
            print(r + "Opci√≥n inv√°lida. Intenta de nuevo." + rst)

# Iniciar la aplicaci√≥n
if __name__ == "__main__":
    menu_principal()
```
